flower.jpg 파일 
- 왼쪽 메뉴바에서 파일 업로드 필요 

```python
from PIL import Image

# 이미지 열기
im = Image.open('flower.jpg') # 이미지 파일은 Image.open()으로 읽어들인다.
im
```

```python
# 이미지 크기 출력
print(im.size)
```

```python
# 이미지 부분 잘라내기
crop_im = im.crop((100,100,200,200)) #im.crop(좌, 상, 우, 하 위치)

crop_im
```

이미지크롭설명.PNG

```python
# 이미지를 파일로 저장
crop_im.save('flower_crop.jpg') 
```

# 실습 과제 작성
- 이미지 크롭 실습

이미지크롭실습.PNG

```python
# 이미지 resize
im_resize = im.resize((200, 150))
im_resize
```

```python
# 이미지 회전
im_rotation = im_resize.rotate(90)
im_rotation
```

```python
from PIL import Image, ImageFilter
# 이미지에 필터를 적용하기 위해서 ImageFilter 추가
```

```python
# 이미지 필터, 블러 필터 사용
im_blur = im_resize.filter(ImageFilter.BLUR) 
#blur 필터를 적용하기 위해 ImageFilter.BLUR 사용
im_blur
```

```python
im_contour = im_resize.filter(ImageFilter.CONTOUR) 
#윤곽선 필터를 적용하기 위해 ImagerFilter.CONTOUR 사용
im_contour
```

```python
#새로운 이미지 파일 만들기
im_new = Image.new('RGB', (500,500), (200,200,200)) 
# 500*500 사이즈 이미지 
# ('RGB', (500,500), (200,200,200))
# (모드, 가로세로크기-px단위, 바탕색)
im_new
```

```python
im_new_bk = Image.new('RGB', (300,500))
# ('RGB', (300,500))
# (모드, 가로세로크기-px단위, 바탕색-기본값 검정)
im_new_bk
```

```python
# 이미지 합치기

im_new = Image.new('RGB', (500,500), (200,200,200)) 
im_new_bk = Image.new('RGB', (100,100)) 

im_new.paste(im_new_bk, (50, 50, 150, 150)) 
# paste() 함수는 이미지를 붙이는 함수. im_new에 im_new_bk를 덧씌우는 동작. 
# 두번째 인수는 위치, 이 값이 붙여진 이미지의 크기와 달라서는 안된다. images do not match
# 왼쪽 상단으로부터 50픽셀씩 떨어진 곳에, 검은색 사각형이 덧 붙여진 결과

im_new
```
